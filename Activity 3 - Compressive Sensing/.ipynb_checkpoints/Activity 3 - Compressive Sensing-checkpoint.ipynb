```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
Fsamp = 8192
dt = 1/Fsamp
t = np.linspace(0,5,Fsamp) 
t
```

```python
f = 5
x = np.sin(2*np.pi*f*t)
plt.plot(x, 'r,')
```

```python
plt.plot(x[:500], 'r,--')
```

```python
F_undersamp = 400
t_undersamp = np.linspace(0,5,F_undersamp ) 
x_undersamp = np.interp(t_undersamp, t, x)
plt.plot(t, x, 'r,')
plt.plot(t_undersamp, x_undersamp, 'bo')
```
